```python
# --- Imports ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --- Load dataset ---

data_path = "../data/melb_data.csv"
df = pd.read_csv(data_path)
# --- Basic inspection ---
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

```python
# --- Data info ---
df.info()
df.describe(include="all")
```

```python
# --- Check for missing values ---
missing = df.isnull().sum().sort_values(ascending=False)
print("Missing values:\n", missing[missing > 0])
```

```python
# --- Target variable distribution ---
plt.figure(figsize=(8,5))
sns.histplot(df['Price'], bins=50, kde=True)
plt.title("Distribution of House Prices in Melbourne")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
# --- Correlation with Price ---
# Only for numeric columns
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()['Price'].sort_values(ascending=False)
print("Top correlations with Price:\n", corr.head(10))
print("\nLowest correlations with Price:\n", corr.tail(10))

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# --- Categorical features quick look ---
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_cols.tolist())
for col in categorical_cols[:5]:  # Just preview first few
    print(f"\n{col} value counts:\n", df[col].value_counts().head())
```

```python
# --- Outlier check (Price vs Rooms, Landsize, Distance, etc.) ---
plt.figure(figsize=(8,5))
sns.scatterplot(x='Rooms', y='Price', data=df, alpha=0.5)
plt.title("Price vs Rooms")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x='Distance', y='Price', data=df, alpha=0.5)
plt.title("Price vs Distance from CBD")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x='Landsize', y='Price', data=df, alpha=0.5)
plt.ylim(0, 3e6)  # cap y for visibility
plt.title("Price vs Land size")
plt.show()
```

```python
# Price heatmap by location
plt.figure(figsize=(10,8))
sns.scatterplot(
    x='Longtitude', 
    y='Lattitude', 
    hue='Price', 
    data=df, 
    palette='coolwarm', 
    alpha=0.5
)
plt.title("Price Heatmap by Location")
plt.show()
```

```python
#Boxplot by number of rooms

plt.figure(figsize=(8,5))
sns.boxplot(x='Rooms', y='Price', data=df)
plt.title("Price Distribution by Number of Rooms")
plt.ylim(0, 3e6)  # cap y-axis for visibility
plt.show()
```

```python
# Boxplot by house area / land size

plt.figure(figsize=(8,5))
sns.boxplot(x=pd.qcut(df['Landsize'], q=5), y='Price', data=df)
plt.title("Price Distribution by Land Size Quantiles")
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot by property type

plt.figure(figsize=(8,5))
sns.boxplot(x='Type', y='Price', data=df)
plt.title("Price Distribution by Property Type")
plt.show()

```
